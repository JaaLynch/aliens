```python
from PIL import Image
import pytesseract
import numpy as np
import pandas as pd
import os

pd.set_option('display.max_rows',1000)
```

```python
df = pd.read_pickle("df.pickle")
```

```python
def search_corpus(df, search_text="alien"):
    df = df[df["file_text"].str.lower().str.contains(search_text.lower(), na=False)]
    print("Document matches: ", df.shape[0])
    return df
```

```python
def view_document(file_name = 'C05516003.tif', folder_name = 'untouched/CIAUFOCD-FULL-UNTOUCHED/'):
    img = Image.open(folder_name+file_name)
    return img
```

### Using the simple functions

```python
# Search for a key word and return document name and corpus
search_corpus(df, "dimension")
```
